```python
import srt
import pandas as pd
import chardet 
from pathlib import Path 

from src.sliding_window import import_hedonometer_as_dict
```

```python
srt_file = Path(
    "Data/subtitles/Fellowship Of The Ring 2001 [Extended Edition] SDH-en.srt"
)
with open(srt_file, "r") as f:
    subtitle_generator = srt.parse(f.read())
subtitles = list(subtitle_generator)
subtitle_df = pd.DataFrame(
    [{"start": s.start, "end": s.end, "content": s.content} for s in subtitles]
)
subtitle_df["start"] = subtitle_df["start"].dt.total_seconds()
subtitle_df["end"] = subtitle_df["end"].dt.total_seconds()
subtitle_df = subtitle_df.sort_values(by="start", ascending=True)
subtitle_df.head()
```

```python
subtitle_df["content"] = subtitle_df["content"].str.replace("\n", " ")
subtitle_df["content"] = subtitle_df["content"].str.replace("<i>", " ")
subtitle_df["content"] = subtitle_df["content"].str.replace("</i>", " ")
```

```python
subtitle_df['wordcount']  = subtitle_df['content'].str.split().str.len()
subtitle_df.to_csv("Data/subtitle-dataframes/fellowship-of-the-ring.csv")
```

```python
trope_file = Path("Data/trope_time_series/Fellowship_of_the_Ring_filled_copy.csv")
with open(trope_file, "rb") as f:
    encoding = chardet.detect(f.read())["encoding"]
trope_df = pd.read_csv(
    "Data/trope_time_series/Fellowship_of_the_Ring_filled_copy.csv", encoding=encoding
)
trope_df["start"] = pd.to_timedelta(
    trope_df["Start Time"], errors="coerce"
).dt.total_seconds()
trope_df["end"] = pd.to_timedelta(
    trope_df["End Time"], errors="coerce"
).dt.total_seconds()
trope_df.head()
```

```python
trope_df["wordcount"] = trope_df["Description"].str.split().str.len()
trope_df["wordcount"].sum()
```

```python
subtitle_df["tropes-active"] = [[] for _ in range(len(subtitle_df))]
subtitle_df["trope-content"] = [[] for _ in range(len(subtitle_df))]
subtitle_df["trope-word-count"] = [0 for _ in range(len(subtitle_df))]


unplaced = []

for trope_idx, trope_row in trope_df.iterrows():
    start_trope = trope_row["start"]
    end_trope = trope_row["end"]
    placed = False

    for sub_idx, sub_row in subtitle_df.iterrows():
        if start_trope < sub_row["end"] and end_trope > sub_row["start"]:
            subtitle_df.at[sub_idx, "tropes-active"].append(trope_row["Trope"])
            subtitle_df.at[sub_idx, "trope-content"].append(trope_row["Description"])
            subtitle_df.at[sub_idx, "trope-word-count"] += trope_row["wordcount"]

            placed = True

    if not placed:
        unplaced.append(
            {
                "start": start_trope,
                "end": end_trope,
                "content": "",
                "tropes-active": [trope_row["Trope"]],
                "trope-content": [trope_row["Description"]],
                "trope-word-count": trope_row["wordcount"],
            }
        )

full_frame = pd.concat([subtitle_df, pd.DataFrame(unplaced)], ignore_index=True)
full_frame = full_frame.sort_values("start").reset_index(drop=True)
full_frame = full_frame.dropna(subset=["start", "end"], how="any", axis=0)
full_frame.tail(30)
```
